```python
#Loaded Titanic dataset using Pandas
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('titanic.csv')
```

```python
#Checked structure and missing values using .info() and .isnull()

df.head()
df.info()
df.shape
```

```python
df.isnull().sum()

```

```python
#Filled missing numeric values (Age) with median and categorical (Embarked) with mode
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
#Removed duplicate records to prevent double counting
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
#Cleaned text columns 'Name' and 'Ticket' by stripping spaces
df['Name'] = df['Name'].str.strip()
df['Ticket'] = df['Ticket'].str.strip()
```

```python
 #Converted 'Survived' and 'Pclass' to integer for calculations
df['Survived'] = df['Survived'].astype(int)
df['Pclass'] = df['Pclass'].astype(int)
```

```python
#Created new analytical features 'Age_Group' and 'Family_Size'
df['Age_Group'] = np.where(df['Age'] < 18, 'Child', 'Adult')

df['Family_Size'] = df['SibSp'] + df['Parch'] + 1
```

```python
#Validated dataset to ensure no nulls remain
df.info()
df.isnull().sum()
```

```python
#Saved cleaned dataset as 'cleaned_data.csv'
df.to_csv('cleaned_data.csv', index=False)
```
